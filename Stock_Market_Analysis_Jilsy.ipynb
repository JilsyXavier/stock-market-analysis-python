**What was the moving average of various stocks?**

```python
#MA of Apple's Stock
ma_day = [20,50,100]
for ma in ma_day:
  column_name ="MA for %s days" % (str(ma))
  df[column_name]=df['Adj Close'].rolling(window=ma,center=False).mean()
df[['Adj Close','MA for 20 days','MA for 50 days','MA for 100 days']].plot(subplots=False,figsize=(12,5),title="MA of APPLE")
```

The moving average (MA) is a simple technical analysis tool that smooths out price data by creating a constantly updated average price.



```python
#MA of Google's Stock
ma_day = [20,50,100]
for ma in ma_day:
  column_name ="MA for %s days" % (str(ma))
  df1[column_name]=df1['Adj Close'].rolling(window=ma,center=False).mean()
df1[['Adj Close','MA for 20 days','MA for 50 days','MA for 100 days']].plot(subplots=False,figsize=(12,5),title="MA of GOOGLE")
```

```python
#Adding a new column Trend whose values based on the daily return
def trend(x):
  if x > -0.015 and x <= 0.015:
    return 'Slight or No change'
  elif x > 0.015 and x <= 0.04:
    return 'Slight Positive'
  elif x < -0.015 and x >= -0.4:
    return 'Slight Negative'
  elif x > 0.04 and x <= 0.06:
    return 'Positive'
  elif x < -0.04 and x >= -0.06:
    return 'Negative'
  elif x > 0.06 and x <= 0.07:
    return 'Among top gainers'
  elif x < -0.06 and x >= -0.07:
    return 'Among top losers'
  elif x > 0.07:
    return 'Bull run'
  elif x <= -0.07:
    return 'Bear drop'
df['Trend']= np.zeros(df['Daily Return'].count()+1)
df['Trend']= df['Daily Return'].apply(lambda x:trend(x))
df.tail(5)
```

**Visualise the trend frequency of Apple in a Pie Chart**

```python
labels=df.Trend.value_counts().index
colors=["olive","orange","hotpink","slateblue","y","lime"]
sizes=df.Trend.value_counts().values
plt.figure(figsize=(5,5))
plt.pie(sizes,labels=labels,colors=colors,autopct="%1.1f%%")
plt.title("Trend Frequency of Apple using a Pie Chart",color="saddlebrown",fontsize=15)
```

What was the correlation between daily returns of different stocks?

```python
#merge the columns of adacent close of all the comoanies
alladj=pd.concat([df['Adj Close'],df1['Adj Close'],df2['Adj Close'],df3['Adj Close']],axis=1)
alladj.columns=['APPLE','GOOGLE','MICROSOFT','AMAZON']
alladj
```

```python
# calculate the percentage change between the adj close of each
pct_df=alladj.pct_change()
pct_df
```

```python
f,ax=plt.subplots(figsize=(12,5))
sns.heatmap(pct_df.corr(),annot=True,linewidths=0.5,linecolor='blue',fmt=".1f",ax=ax)
```
